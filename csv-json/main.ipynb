```python
import csv
import json
import uuid
import bcrypt
from datetime import datetime

# File paths
input_csv_path = 'D:\\Next JS\\examiner-app\\college_data.csv'
output_json_path = 'colleges.json'

# Read the CSV file and process each row
def process_csv():
    colleges = []
    with open(input_csv_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile, delimiter=';', skipinitialspace=True)
        for row in reader:
            print(row)
            # Generate a unique identifier for each college
            college_id = str(uuid.uuid4())
            # Hash the password
            hashed_password = bcrypt.hashpw(row['Password'].encode('utf-8'), bcrypt.gensalt())
            # Create a college dictionary
            college = {
                '_id': college_id,
                'name': row['College Name '],
                'collegeId': row['College Code'].strip(),
                'password': hashed_password.decode('utf-8'),  # Store as a string
                'status': 'active',
                'plainPassword': row['Password'],  # Consider omitting or securing this field
                'createdAt': datetime.now().isoformat()  # Timestamp
            }
            colleges.append(college)
    
    # Write to JSON file
    with open(output_json_path, 'w', encoding='utf-8') as jsonfile:
        json.dump(colleges, jsonfile, ensure_ascii=False, indent=2)

    print('JSON file has been written successfully.')

# Run the function
process_csv()
```
